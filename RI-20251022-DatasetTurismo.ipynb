```python
import numpy as np
def get_tf(doc):
    vec =[]
    for word in vocab:
        cont = doc.lower().replace('.', '').replace(',', '').split().count(word)
        vec.append(cont)
    return np.array(vec)
```

```python
x = get_tf(corpus[0])
x
```

```python
corpus_vec = []
for doc in corpus:
    doc_vec = get_tf(doc)
    corpus_vec.append(doc_vec)
```
